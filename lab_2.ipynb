```python
# Задание task_02_17
#Задание: решите задачу № 2.17, организовав бесконечный цикл, который бы прерывался при выполнении условия, используя оператор break .
# Выполнила: Михалева Полина Вячеславовна
# Группа: ЦИБ-251

nums_sum = 0
nums_count = 0

while True:
    if nums_count == 0:
        text = "Введите первое число: "
    elif nums_count == 1:
        text = "Введите второе число: "
    elif nums_count == 2:
        text = "Введите третье число: "
    else:
        text = "Введите " + str(nums_count + 1) + "-е число: "

    num = int(input(text))

    if num == 0:
        break

    nums_sum += num
    nums_count += 1

print("Сумма =", nums_sum)
print("Количество =", nums_count)
```
