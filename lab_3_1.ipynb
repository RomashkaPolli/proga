```python
# Задание task_03_01_17.
#Дано натуральное число. Напишите рекурсивные функции для определения:суммы цифр числа; количества цифр в числе.
# Выполнила: Михалева Полина Вячеславовна
# Группа: ЦИБ-251

def digits_sum(value):
    """Вернуть сумму цифр числа 'value'. Рекурсивная реализация.

    Параметры:
        - value: натуральное число.

    Результат:
        сумма цифр числа.
    """
    if value < 10:
        return value
    else:
        return value % 10 + digits_sum(value // 10)

def количество_цифр(value):
    """Вернуть количество цифр в числе 'value'. Рекурсивная реализация.

    Параметры:
        - value: натуральное число.

    Результат:
        количество цифр в числе.
    """
    if value < 10:
        return 1
    else:
        return 1 + количество_цифр(value // 10)

print(digits_sum(12345))
print(количество_цифр(12345))
```
